```python
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.core.display import HTML
```

```python
# our animation: a number that counts up

fig, ax = plt.subplots()

t = ax.text(0.5, 0.5, "", size=72, va="center", ha="center")

def draw_frame(frame_num):
    t.set_text(str(frame_num))

anim = FuncAnimation(fig, draw_frame, 10, interval=1000)
html = anim.to_html5_video()
plt.close()
HTML(html)
```

```python
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.core.display import HTML, display
from time import time

fig, ax = plt.subplots(figsize=(4,4))

circ = plt.Circle((0, 0), 0.1, facecolor="red", zorder=3)
ax.add_artist(circ)
rect = plt.Rectangle((0, 0.3), 1, 0.1, facecolor="black", zorder=2)
ax.add_artist(rect)

vid_seconds = 2
fps = 5 # frames per second
frame_count = vid_seconds*fps

def draw_frame(frame_num):
    percent = frame_num / frame_count # 0 and 1
    circ.center = (0.5, 1-percent)

debug_frame = None # None means make an animation instead of debugging

if debug_frame == None:
    anim = FuncAnimation(fig, draw_frame, frames=frame_count, interval=1000/fps)
    t0 = time()
    html = anim.to_html5_video()
    t1 = time()
    print("Seconds to generate:", t1-t0)
    plt.close(fig)
    display(HTML(html))
else:
    draw_frame(debug_frame)
```
